# NBA Top 3 Players

This file will explore the comparisons between Lebron James, Michael Jordan, and Kobe Bryant. I have gathered season stats for each of the three players and used the data to analyze their relative performances.

#### Lebron James-Small Forward

Generally thought of as the most versatile player. Has the ability to dive to the basket and finish strong or with a skillful layup. Has developed and extremely effective post game. Jumpshot is hit-or-miss, with noticable struggles from the free throw line. Extremelly athletic, lockdown defender, and can pass like an elite point guard.

#### Michael Jordan-Shooting Guard

Explosive, high-flying player who can score at will. Great post game with signiture fadeway jumpshot. Unmatched athletisicm that he used to perform insane dunks and blow by defenders. Known for winning 6 NBA championships and and making clutch shots in big games. 

#### Kobe Bryant-Shooting Guard

Thought of as the second coming of Jordan. Scoring monster who is always in attack mode. Has deadly jumpshot mixed with elite athletisicm, great defense, and a relentless competitive spirit

I imported season stats from basketball-reference.com into Excel sheets. I removed season totals where the player played less that 20 games to keep the data precise. The Excel sheets were saved at CSV files and then imported into the dataframes 'lebron', 'jordan', and 'kobe'.

```r
#Imports stats into dataframe
lebron <- read.csv('Lebron.csv')
jordan <- read.csv('Jordan.csv')
kobe <- read.csv('Kobe.csv')
```

## Analysis

I generated graphs of points per game, rebounds per game, steals per game, blocks per game, field goal percentage, 
3 point percentage, and turnovers per game. 

Lebron dominates the most of the 8 categories. His 6'8', 240lb frame goves him a huge advantage in rebounding. He is averaging a block more per game than Jordan, and two more than Kobe. His passing skills are exceptional, as he averaged almost 2 more assister per game than Jordan and three more than Kobe.  He is definitely a more versatile player that can do many different things on the court. Though he is the most consistent, I cannot ignore that Lebron does have the most turnovers. He even has averaged almost 4 turnoves a game twice in his career. I do believe this is because he attempting more passes per game and therefore has a higher chance of turning the ball over. 

Jordan was the most successful scorer, having 2 seasons where he averages 35 PPG or more. This has only been done a few times
in the history of the NBA, so the fact that he has done it twice is astounding. He also has the most seasons averaging 30 points or more with 8 seasons, followed by Kobe with 3 and Lebron with 2. His average field goal percentage is just below 50%, proving he was also scoring efficiently. Jordan also has the most steals and blocks, suggesting he was the best defender of the three.

Kobe's stats are disappointing because they are clearly the lowest for rebounds, assists, steals and blocks. While he is generally thought to be the best shooter of the three, his field goal percentage is the lowest, while his three point percentage looks slightly worse than Lebron's. He has been known to shoot ill-advised shots with extremelly high difficulty on a regular basis. These shots can definitely be going in some nights, but in the long term hurt his overall FG percentage. On his decline, he even shot 37.3% and 35.8%, the lowest two percentages for any season of the three players.

```r
#Creates plot layout
par(mfrow = c(2, 4))

#Generates boxplots of stats
boxplot(lebron$PTS, jordan$PTS, kobe$PTS, main = 'Points Per Game', ylab = 'Points', names = c('Lebron', 'Jordan', 'Kobe'), col = 3, boxwex = 0.6, cex.main = 1.2, cex.axis = 0.6)
boxplot(lebron$TRB, jordan$TRB, kobe$TRB, main = 'Rebounds', ylab = 'Rebounds', names = c('Lebron', 'Jordan', 'Kobe'), col = 4, boxwex = 0.6, cex.main = 1.2, cex.axis = 0.6)
boxplot(lebron$AST, jordan$AST, kobe$AST, main = 'Assists', ylab = 'Assists', names = c('Lebron', 'Jordan', 'Kobe'), col = 5, boxwex = 0.6, cex.main = 1.2, cex.axis = 0.6)
boxplot(lebron$STL, jordan$STL, kobe$STL, main = 'Steals', ylab = 'Steals', names = c('Lebron', 'Jordan', 'Kobe'), col = 6, boxwex = 0.6, cex.main = 1.2, cex.axis = 0.6)

boxplot(lebron$BLK, jordan$BLK, kobe$BLK, main = 'Blocks', ylab = 'Blocks', names = c('Lebron', 'Jordan', 'Kobe'), col = 7, boxwex = 0.6, cex.main = 1.2, cex.axis = 0.6)
boxplot(lebron$FGp, jordan$FGp, kobe$FGp, main = 'FG Percentage', ylab = 'Percent', names = c('Lebron', 'Jordan', 'Kobe'), col = 8, boxwex = 0.6,cex.main = 1.2, cex.axis = 0.6)
boxplot(lebron$X3Pp, jordan$X3p, kobe$X3Pp, main = '3 Point Percentage', ylab = 'Percent', names = c('Lebron', 'Jordan', 'Kobe'), col = 0, boxwex = 0.6, cex.main = 1.2, cex.axis = 0.6)
boxplot(lebron$TOV, jordan$TOV, kobe$TOV, main = 'Turnovers', ylab = 'Turnovers', names = c('Lebron', 'Jordan', 'Kobe'), col = 10, boxwex = 0.6, cex.main = 1.2, cex.axis = 0.6)
```

## Three Point Shooting

Kobe would never be accused of having a poor jumpshot since he was known to knock down difficult jumpers. Both Lebron and Jordan came into the league with one noticable flaw - lack of a consistent jumpshot. Jordan would improve his jumpshot exponentially and went from shooting and awful 17% in his rookie year to 42% in 1995-1996. Lebron's shooting improved slightly and he has been very consistent around 32.18%-36.00% per season. However, the 3 point percentage stat shows that Lebron actually has the highest average 3 point percentage. His boxplot range is also the smallest, showing the is the most consistent long range shooter. 


```r
#Generate summary of 3 point percentages
summary(lebron$X3Pp)
summary(jordan$X3p)
summary(kobe$X3Pp)
```

I used more data visualization techniques to examine the 3 point shooting. The top row of graphs shows the difference between the 3 point shots attempted versus 3 point shots made. The gap between the red and green lines relates to the 3 point percentage, where a large gap means lower shooting percentage and a small gap means higher shooting percentage. The bottom row of graphs shows 3 point percentage for each season.

Jordan has the lowest few recorded percentages of all of the three. He seriously struggled in his first 3 seasons, unable to average more than 18.2 percent in each. He later improved drastically enough to post the highest percentage of 42.7% in at age 32. This was after he had only played the last 17 games of the prior season, and had been retired for the two seasons before that. I believe that Jordan wasn't only just playing minor league baseball and that he was still working on his jumpshot. In those 17 games he played after his return, he shot and improved 50.0% from the three point line. He then went on to average 33.22% for the next 3 seasons after only averaging 24.58% for his first 7 seasons.

Lebron and Kobe's scatterplot spreads are both very precise. However, one thing I noticed was that Lebron's percentage increases with time, while Kobe's decreases. I believe this is a reflection of the way defenders guard each player. Lebron is know as a "drive first then shoot", so defenders are told to back up off him and force him to hit the open jumpshot. This called an uncontested shot, and is the easiest 3 point shot that can be taken. Kobe is a "shoot first then drive" so defenders will be close to his body, guarding any jumpshot that he will attempt. This forces him to take tough fadeaways, shoot off the dribble, and take contested shots. Since these shots have much higher difficulty, it makes sense that Kobe's percentage is lower even though he is actually a better 3 point shooter. 

```r
par(mfrow = c(2, 3))

plot(lebron$Age, lebron$X3P, main = 'Lebron 3 pointers', xlab = 'Age', ylab = 'Shots', type = 'l', ylim = c(0, 8), col = 3)
lines(lebron$Age, lebron$X3PA, col = 2)
legend(19, 8, c('Made', 'Attempted'), lty=c(1,1), col=c(3, 2), cex = 0.85)

plot(jordan$Age, jordan$X3P, main = 'Jordan 3 pointers', xlab = 'Age', ylab = 'Shots', type = 'l', ylim = c(0, 8), col = 3)
lines(jordan$Age, jordan$X3PA, col = 2)
legend(21, 8, c('Made', 'Attempted'), lty=c(1,1), col=c(3, 2), cex = 0.85)

plot(kobe$Age, kobe$X3P, main = 'Kobe 3 pointers', xlab = 'Age', ylab = 'Shots', type = 'l', ylim = c(0, 8), col = 3)
lines(kobe$Age, kobe$X3PA, col = 2)
legend(18, 8, c('Made', 'Attempted'), lty=c(1,1), col=c(3, 2), cex = 0.85)

plot(lebron$Age, lebron$X3Pp, pch = 25, col = 2, bg = 2, main = 'Lebron 3 Percentage', xlab = 'Age', ylab = 'Percent', ylim = c(0.1, 0.5))
plot(jordan$Age, jordan$X3p, pch = 25, col = 2, bg = 2, main = 'Jordan 3 Percentage', xlab = 'Age', ylab = 'Percent', ylim = c(0.1, 0.5))
plot(kobe$Age, kobe$X3Pp, pch = 25, col = 2, bg = 2, main = 'Kobe 3 Percentage', xlab = 'Age', ylab = 'Percent', ylim = c(0.1, 0.5))
```

## Defensive Play

All three players are all-time great defenders. Kobe and Jordan have been named to the NBA All-Defensive 1st Team 9 times, and Lebron has been named 5 times. To analyze the defensive abilities of each player, I generated more indepth graphs of steals and blocks.

Jordan clearly dominates in steals, averaging 2.34 steals per game to Lebron's 1.65 and Kobe's 1.41. This is a surprising difference, considering all three are elite defenders. I believe his ability to create turnovers stems from his high basketball IQ. He is able to read players well enough to predict their next move and steal the ball. Lebron and Kobe's steals per game are also very high, so Jordan averaging almost a steal more than both of them is impressive.

As far as blocks, you would think Lebron would win this one because he is more of a low-post player who can block many shots with his supreme athletisicm. Surprisingly, Jordan also wins this category. As a shooting guard, it is very uncommon to have anywhere near 1.5 blocks per game. Jordan has done this twice in his career. However, his average gradually decreases after those two years. This could be due to improved team defense from the play of Scottie Pippen and Dennis Rodman. With improved team defense, Jordan would not have to be as prevalent on defense so that hes could focus on offense. This is backed by his increase in points per game, 3 point percentage, and  field goal percentage in the next few seasons.

```r
par(mfrow = c(2, 3))

barplot(lebron$STL, main = "Lebron", cex.main = 1.4, font.main = 1, xlab = '', ylab = 'Steals', cex.lab = 1.2, ylim = c(0, 3.5), col = 3)
barplot(jordan$STL, main = "Jordan", cex.main = 1.4, font.main = 1, xlab = '', ylab = 'Steals', cex.lab = 1.2, ylim = c(0, 3.5), col = 5)
barplot(kobe$STL, main = "Kobe", cex.main = 1.4, font.main = 1, xlab = '', ylab = 'Steals', cex.lab = 1.2, ylim = c(0, 3.5), col = 4)

barplot(lebron$BLK, main = "Lebron", cex.main = 1.4, font.main = 1, xlab = '', ylab = 'Blocks', cex.lab = 1.2, ylim = c(0, 1.7), col = 3)
barplot(jordan$BLK, main = "Jordan", cex.main = 1.4, font.main = 1, xlab = '', ylab = 'Blocks', cex.lab = 1.2, ylim = c(0, 1.7), col = 5)
barplot(kobe$BLK, main = "Kobe", cex.main = 1.4, font.main = 1, xlab = '', ylab = 'Blocks', cex.lab = 1.2, ylim = c(0, 1.7), col = 4)
```
